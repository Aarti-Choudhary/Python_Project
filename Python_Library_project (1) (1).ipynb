### Exploratory Data Analysis (EDA)

###  Attrition by department

```python
plt.figure(figsize=(8, 6))  # Adjust the width and height as needed
plt.subplot(1, 2, 1)
sns.countplot(x='Department', hue='Attrition', data=dataset, linewidth = 4)
plt.title('Attrition by Department')
plt.xticks(rotation=45, ha = "right")  ## Rotated x-axis labels to horizontal line since it was overlapping
plt.show()

#Department with the highest attrition rate: Research and Development
```

### Attrition Distribution Across Different Departments

```python
attrition_distribution = dataset.groupby('Department')['Attrition'].value_counts().unstack()  # Unstacking the 'Group' level of the index
attrition_distribution.plot(kind='bar', stacked=True, figsize=(10, 6))  # kind is a parameter which species which bar you want to generate inseated of kind we can take x,y as well
plt.title('Attrition Distribution Across Different Departments')   # stacked if you want to show the bar plot on top instead of showing side by side we can use stacked = t
plt.xlabel('Department')
plt.ylabel('Count of Employees')
plt.legend(title='Attrition')
plt.xticks(rotation=45, ha = "right")
plt.show()
```

### JobSatisfaction vs Attrition

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=dataset, x='JobSatisfaction', hue='Attrition')
plt.title('Job Satisfaction vs Attrition')
plt.xlabel('Job Satisfaction Level')
plt.ylabel('Count')
plt.legend(title='Attrition')
plt.show()
```

### WorkLifeBalance vs Attrition

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=dataset, x='WorkLifeBalance', hue='Attrition', color = 'red')  # hue is a paramtetr
plt.title('Work-Life Balance vs Attrition')
plt.xlabel('Work-Life Balance Level')
plt.ylabel('Count')
plt.legend(title='Attrition')
plt.show()
```

### Attrition rate by marital status.

```python
attrition_by_marital_status = dataset.groupby('MaritalStatus')['Attrition'].value_counts(normalize=True).unstack()

# Plot attrition rate by marital status
plt.figure(figsize=(8, 6))
attrition_by_marital_status.plot(kind='bar', stacked=True, color=['yellow', 'm'])

# Add labels and title
plt.title('Attrition Rate by Marital Status')
plt.xlabel('Marital Status')
plt.ylabel('Attrition Rate')

# Show plot
plt.tight_layout()
plt.show()
```

### Monthly income by department


```python
# Boxplot 
plt.figure(figsize=(10, 6))
sns.boxplot(x='Department', y='MonthlyIncome', data=dataset , color = 'hotpink')
plt.title('MonthlyIncome by Department')
plt.xlabel('Department')
plt.ylabel('Monthly Income')
plt.show()

##sales has the highest income
```

### Histogram of Age distribution

```python

plt.figure(figsize=(8, 6))
sns.histplot(dataset['Age'], bins=20, kde=True , color = 'orange')   #bins is a interval ranges which divides data for the purpose of analsis
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python

data_filtered = dataset[['Department', 'JobSatisfaction', 'MonthlyIncome']] # Data Preparation

# Comparison Analysis: Calculated means for each department
department_comparison = data_filtered.groupby('Department').mean().reset_index()

plt.figure(figsize=(12, 6))

# Bar plot for Job Satisfaction
plt.subplot(1, 2, 1)
sns.barplot(x='Department', y='JobSatisfaction', data=data_filtered, ci=None, color = 'g')  #confidence interval
plt.title('Mean Job Satisfaction by Department')

# Bar plot for Monthly Income
plt.subplot(1, 2, 2)
sns.barplot(x='Department', y='MonthlyIncome', data=data_filtered, ci=None)
plt.title('Mean Monthly Income by Department')

plt.show()
```

### Distributions Of distnace from home

```python
distances_from_home = dataset['DistanceFromHome']

# Visualize the distribution using a histogram
plt.figure(figsize=(8, 6))
sns.histplot(distances_from_home, bins=20, kde=True, color='skyblue')
plt.title('Distribution of Distances from Home')
plt.xlabel('Distance from Home')
plt.ylabel('Frequency')
plt.show()
```

### Distribution of Education Levels Among Employees

```python
education_levels = dataset['Education']

# Visualize the distribution using a countplot
plt.figure(figsize=(8, 6))
sns.countplot(x=education_levels, palette='viridis')   #"Viridis" refers to a colormap in data visualization
plt.title('Distribution of Education Levels Among Employees')
plt.xlabel('Education Level')
plt.ylabel('Frequency')
plt.show()
```

### Distribution of employees across different education fields.


```python
education_field = dataset['EducationField']

education_field_counts = education_field.value_counts() # here i have counted the frequency of each education field

# Visualize the distribution using a bar plot or count plot
plt.figure(figsize=(8, 6))
sns.countplot(x='EducationField', data=dataset, order=education_field_counts.index, palette='viridis')
plt.title('Distribution of Employees Across Different Education Fields')
plt.xlabel('Education Field')
plt.ylabel('Number of Employees')
plt.show()
```

```python
environment_satisfaction = dataset['EnvironmentSatisfaction']
job_satisfaction = dataset['JobSatisfaction']
satisfaction_levels = pd.concat([environment_satisfaction, job_satisfaction])
plt.figure(figsize=(8, 6))
sns.histplot(satisfaction_levels, bins=4, kde=True, color='m')  # kde is density estimation  if i want 4 segregated bins can  write bins = 8
plt.title('Distribution of Satisfaction Levels')
plt.xlabel('Satisfaction Level')
plt.ylabel('Frequency')
plt.xticks([1, 2, 3, 4], ['Low', 'Medium', 'High', 'Very High'])  #  1-4 is  for satisfaction levels
plt.show()
```

### Distribution of Marital Statuses Among Employees

```python
marital_status = dataset['MaritalStatus']

marital_status_counts = marital_status.value_counts()   ## Count the occurrences of each marital status

plt.figure(figsize=(8, 6)) # Creating the plot

# Plot the distribution of marital statuses
sns.barplot(x=marital_status_counts.index, y=marital_status_counts.values, palette='viridis')

# Add labels and title
plt.title('Distribution of Marital Statuses Among Employees')
plt.xlabel('Marital Status')
plt.ylabel('Frequency')

plt.show()

#Note -->.the insights gained from analyzing marital status distributions to inform strategic decision-making is the married employees
```

### Distribution of years spent at the company by employees.

```python
years_at_company = dataset['YearsAtCompany']

# Create the plot
plt.figure(figsize=(8, 6))

# Plot the distribution of years spent at the company
sns.histplot(years_at_company, bins=20, kde=True, color='hotPink')

# Add labels and title
plt.title('Distribution of Years Spent at the Company by Employees')
plt.xlabel('Years at Company')
plt.ylabel('Frequency')

# Show the plot
plt.show()
```
